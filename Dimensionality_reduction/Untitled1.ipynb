```python
def is_countercolwise(subpolygon):  
    area = 0
    counterclokwise=False
    for index,_ in enumerate(sub_polygon):
        second_index=(index+1)%len(sub_polygon)
        area+=sub_polygon[index][0]*sub_polygon[second_index][1]
        area-=sub_polygon[second_index][0]*sub_polygon[index][1]
    if area/2<0:
        counterclokwise=False
    else:
        counterclokwise=True
    return counterclokwise
        
  
```

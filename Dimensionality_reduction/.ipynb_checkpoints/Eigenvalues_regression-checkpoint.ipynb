```python
%matplotlib inline

import math
import matplotlib.pyplot as plt
import matplotlib.tri as tri

import numpy as np

from Deformation_gradient import get_eigenvalues


import torch
import torch.optim as optim
import torch.nn as nn
from torch.autograd import Variable
import torch.nn.functional as Func


from IPython.core.debugger  import Tracer

```

```python

from Deformation_gradient import get_eigenvalues

nb_of_triangles=5005
Y=np.empty([nb_of_triangles,2])
X=np.empty([nb_of_triangles,5])
for i in range(nb_of_triangles):
    triangle=np.random.normal(0,1,(3,2))
    #triangle=np.array([[-1,0],[1,0],[0,sqrt(3)]])
    triangle=np.hstack((triangle,[[0],[0],[0]]))
    
    #fig=plt.figure()
    #ax=fig.add_subplot(111)
    #plt.triplot(triangle[:,0],triangle[:,1])
    
    Tracer()() 
    eig=get_eigenvalues(triangle)
    
    #X[i]=F.flatten()
    
    #Y[i]=np.array(eig)

eig    
    
    
```

```python
# Convert to pytorch tennsors

x_tensor=torch.from_numpy(X).type(torch.FloatTensor)
y_tensor=torch.from_numpy(Y).type(torch.FloatTensor)

# Convert to pytorch variables
x_variable=Variable(x_tensor,requires_grad=False)
y_variable=Variable(y_tensor,requires_grad=False)
```

```python
class Net(torch.nn.Module):
    def __init__(self, n_feature, n_hidden, n_output):
        super(Net, self).__init__()
        self.fc1 = torch.nn.Linear(n_feature, n_hidden)   # hidden layer
        #self.fc2 = torch.nn.Linear(n_hidden, n_hidden)   # hidden layer
        self.fc3 = torch.nn.Linear(n_hidden, n_hidden)   # hidden layer
        self.fc4 = torch.nn.Linear(n_hidden, n_output)   # output layer

    def forward(self, x):
        x = Func.relu(self.fc1(x))      # activation function for hidden layer
        #x = Func.relu(self.fc2(x))
        x = Func.relu(self.fc3(x))

        x = self.fc4(x)

        return x
```

```python
# Set hyperparameters #
net = Net(n_feature=4, n_hidden=40, n_output=2)     # define the network
print(net)  # net architecture

optimizer = torch.optim.SGD(net.parameters(), lr=0.0001,momentum=0.9)
loss_func = torch.nn.MSELoss()  


```

```python
# Train the network #
for t in range(1200):
    out = net(x_variable)                 # input x and predict based on x
    loss = loss_func(out, y_variable)     # must be (1. nn output, 2. target), the target label is NOT one-hotted

    optimizer.zero_grad()   # clear gradients for next train
    loss.backward()         # backpropagation, compute gradients
    print(t,loss.data[0])
    optimizer.step()        # apply gradients
```
